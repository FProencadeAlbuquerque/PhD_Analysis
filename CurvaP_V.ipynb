```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
phi=0
```

## $ p_{max}=\dfrac{P_{max}X}{E^2}=\dfrac{ \cos \phi}{2(1 + \sin \phi) }$,  

## $v_{crit}=\dfrac{|\dot V_{max}|}{|\dot E|}=\dfrac{1}{\sqrt{2} \cdot \sqrt{1+\sin \phi}}$

```python
pmax=np.cos(phi)/(2*(1+np.sin(phi)))

vcrit=(1/np.sqrt(2))/(np.sqrt(1+np.sin(phi)))
```

```python
p=np.linspace(0,pmax,100)
```

## $v=\sqrt{-\dfrac{2p \tan \phi -1}{2} \pm \dfrac{\sqrt{(2p \tan \phi -1)^2-4p^2\sec^2 \phi}}{2}}=\sqrt{\dfrac{f_1}{2} \pm \dfrac{f_2}{2}}$

```python
f1=-(2*p*np.tan(phi)-1)

f2=np.sqrt((2*p*np.tan(phi)-1)**2-4*p**2*1/(np.cos(phi)*np.cos(phi)))
```

## $v_1=\sqrt{\dfrac{f_1}{2} + \dfrac{f_2}{2}}$

## $v_2=\sqrt{\dfrac{f_1}{2} - \dfrac{f_2}{2}}$

```python
v1=np.sqrt(f1/2+f2/2)

v2=np.sqrt(f1/2-f2/2)
```

```python
fig, ax = plt.subplots(figsize = (13,8))

axis_font = {'fontname':'sans-serif', 'size':'25'}

plt.xlim([0, 0.5])

plt.xlabel(r'$p \ (pu)$',**axis_font)
plt.ylabel(r'$v \ (pu)$',**axis_font)
plt.yticks(fontsize=25,**axis_font)
plt.xticks(fontsize=25,**axis_font)
plt.plot(p,v1,p,v2,linewidth=3.0)

plt.show()
```
