```python
import numpy as np
```

```python
npontos=101

xn=np.ones((npontos,1))

Lambda=np.linspace(0,1,npontos)
    
Delta_lambda=Lambda[1]-Lambda[0]

xn[0]=-1
```

```python
for j in range(npontos-1):
  xn[j+1]=xn[j]-Delta_lambda*(1/(3*xn[j]**2-3*Lambda[j]))*(-3*xn[j])

  sol=np.ones((20,1))

  sol[0]=xn[j+1]

  for i in range(19):

     sol[i+1]=sol[i]-((sol[i]**3-3*Lambda[j+1]*sol[i]+1)/(3*sol[i]**2-3*Lambda[j+1]))

  xn[j+1]=sol[19]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (13,8))

axis_font = {'fontname':'Times New Roman', 'size':'25'}

plt.xlim([0,1])

plt.xlabel(r'$\lambda$',**axis_font)
plt.ylabel(r'$x$',**axis_font)
plt.yticks(fontsize=25,**axis_font)
plt.xticks(fontsize=25,**axis_font)
plt.plot(Lambda,xn,linewidth=3.0)

plt.show()
```
