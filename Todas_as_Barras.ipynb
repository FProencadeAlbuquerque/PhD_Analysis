```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Análise dos algoritmos:

*   Extra Tree Regressor
*   Random Forest Regressor
*   Decision Tree Regressor

# Em condições normais





```python
bus=np.array([4,5,9,10,11,12,13,14])

aux=np.linspace(0,20,100)

MAPE_RF=np.array([1.0460408915807515,1.4163282522637302, 1.4302617786830119,1.5186141529478623,1.946224306997004,1.2701550401570094,1.86412176339142,2.3375])

media_RF=np.array(MAPE_RF).mean()

media_RF_str=str(f'{media_RF:.4f}')

media_RF_vec=[]

for i in range(len(aux)):
  media_RF_vec.append(media_RF)

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (14,8))

axis_font = {'fontname':'sans-serif', 'size':'23'}

plt.bar(bus,MAPE_RF)

plt.plot(aux,media_RF_vec,'b--',linewidth=3.0, label=r'$\overline{MAPE}=$'+media_RF_str+'%')

plt.title('Performance usando Random Forest',**axis_font)
plt.xlabel('Barras PQ do Sistema',**axis_font)
plt.ylabel('MAPE (%)',**axis_font)
plt.yticks(fontsize=23,**axis_font)
plt.xticks([4,5,9,10,11,12,13,14],fontsize=23,**axis_font)
plt.xlim((3,15))
plt.legend(fontsize=21) 

#plt.savefig('/content/drive/MyDrive/AlgoritmosDOC/Barras_Margem_Estabilidade/TodasCon_RF_2.eps', format='eps')

plt.show()
```
