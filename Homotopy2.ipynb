```python
import numpy as np
```

```python
npontos=1001

xn1=np.ones((npontos,1))

Lambda=np.linspace(1,10,npontos)
    
Delta_lambda=Lambda[1]-Lambda[0]

tam=len(xn1)

xn1[tam-1]=5.46048
```

```python
for j in range(npontos-1):

  xn1[tam-j-2]=xn1[tam-j-1]-Delta_lambda*(1/(3*xn1[tam-j-1]**2-3*Lambda[tam-j-2]))*(-3*xn1[tam-j-1])

  sol=np.ones((20,1))

  sol[0]=xn1[tam-j-2]

  for i in range(19):

    sol[i+1]=sol[i]-((sol[i]**3-3*Lambda[tam-j-2]*sol[i]+1)/(3*sol[i]**2-3*Lambda[tam-j-2]))

  xn1[tam-j-2]=sol[19]
```
